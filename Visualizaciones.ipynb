# Generar datos aleatorios


```python
# Inicializa en 0
x = 0
y1 = 0
y2 = 0

x_list = []
y_list1 = []
y_list2 = []
# Aumenta o disminuye aleatoriamente
for i in range(15):
  x = x + 1                       # x += 1
  y1 = y1 + random.randint(-6, 8) # y1 += random.randint(-6, 8)
  y2 = y2 + random.randint(-5, 6) # y2 += random.randint(-5, 6)
  x_list.append(x)
  y_list1.append(y1)
  y_list2.append(y2)
```

# Crear gráfica de los datos generados previamente

```python
plt.plot(x_list, y_list1)
plt.plot(x_list, y_list2)
```

Para diferenciar una gráfica de la otra utilizamos el parámetro label y la función legend

```python
plt.plot(x_list, y_list1, label = 'Total 1')
plt.plot(x_list, y_list2, label = 'Total 2')
plt.title("Datos generados aleatoriamente")
plt.xlabel("Tiempo")
plt.ylabel("Muestras")
plt.legend(loc = 'upper left')
plt.tight_layout()
```

```python
# Crear una Figura y un array de subplots con 2 filas y 2 columnas
fig, ax = plt.subplots(2, 2, figsize=(16,10))
fig.tight_layout(pad = 5.0)

ax[0, 0].hist(df["Age"], color = "#887")
ax[0, 0].set_xlabel("Rango de edades")
ax[0, 0].set_ylabel("Cantidad de personas")
ax[0, 0].set_title("Distribución de edades")

df_female = df[df["Gender"] == "Female"]
df_male = df[df["Gender"] == "Male"]
#df_male['Gender'].value_counts()

ax[0, 1].boxplot([df_female["YearsAtCompany"], df_male["YearsAtCompany"]])
ax[0, 1].set_xticklabels(['Female','Male'])
ax[0, 1].set_ylabel("YearsAtCompany")
ax[0, 1].set_title("Lealtad")


my_colors = 'rgbkymcrg'
ax[1, 0].bar(df["JobRole"].value_counts().index, df["JobRole"].value_counts(), color=my_colors)
ax[1, 0].set_xticklabels(df["JobRole"].value_counts().index, rotation=90)
ax[1, 0].set_title("Cantidad de personas por rol")

img = ax[1, 1].scatter(df_sorted_age["Age"], df_sorted_age["MonthlyIncome"], c = df_sorted_age['JobSatisfaction'], cmap='BrBG')
ax[1, 1].set_xlabel("Edad")
ax[1, 1].set_ylabel("MonthlyIncome")
ax[1, 1].set_title("Relación de edad-salario-satisfacción")
cbar = fig.colorbar(img)
cbar.ax.set_ylabel('JobSatisfaction')

plt.show()
```

# 2 Ejemplos de la documentación para gráficas 3D
https://matplotlib.org/2.0.2/mpl_toolkits/mplot3d/tutorial.html

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.gca(projection='3d')

# Make the grid
x, y, z = np.meshgrid(np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.8))

# Make the direction data for the arrows
u = np.sin(np.pi * x) * np.cos(np.pi * y) * np.cos(np.pi * z)
v = -np.cos(np.pi * x) * np.sin(np.pi * y) * np.cos(np.pi * z)
w = (np.sqrt(2.0 / 3.0) * np.cos(np.pi * x) * np.cos(np.pi * y) *
     np.sin(np.pi * z))

ax.quiver(x, y, z, u, v, w, length=0.1, normalize=True)

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure()
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
import folium
import pandas as pd
```

```python
df = pd.read_csv('denue_inegi_19_.csv', engine='python')
```

```python
df.head()
```

```python
subset_of_df = df.sample(n=500)
```

```python
m = folium.Map(
              location=[subset_of_df['latitud'].mean(),subset_of_df['longitud'].mean()],
              zoom_start=6
              )
m
```

```python
for row in subset_of_df.itertuples():
  m.add_child(folium.Marker(location=[row.latitud, row.longitud], popup=str(row.id)))
```
